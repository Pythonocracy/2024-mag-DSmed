***Различия между Fpz-Cz и Pz-Oz***

Разница между этими парами электродов состоит в их расположении и функциональной роли:

**Fpz-Cz**:
Fpz регистрирует дельта-активность, особенно важную для анализа глубоких стадий сна, тогда как Cz помогает оценить двигательную активность и функции центральной нервной системы.
Эти электроды находятся ближе к передней части мозга и отвечают за различные аспекты обработки сенсорной информации и координации движений.

**Pz-Oz**:
Pz регистрирует альфа-активность и реагирует на стимулы зрения, что делает его полезным для анализа зрительных реакций и состояний покоя.
Oz также регистрирует быструю активность, связанную с когнитивными функциями и вниманием.
Они расположены дальше назад и отвечают преимущественно за обработку зрительной информации и координацию внимания.





ЭОГ (электроокулография) – это метод измерения движения глазных яблок путем регистрации электрических потенциалов, вызванных изменением поляризации сетчатки глаза при перемещении взгляда. Это неинвазивный метод, позволяющий отслеживать направление и скорость движения глаз.

ЭМГ (электромиография) – метод регистрации электрической активности мышц, основанный на измерении биопотенциалов, возникающих в мышцах при их возбуждении. Это важный инструмент в медицинской диагностике, физиологии и спортивной науке, позволяющий изучать работу мышц и оценивать их состояние. В нашем случае даны измерения подбородочной ЭМГ.



```python
# Функция для вейвлет-преобразования временных рядов
def wavelet_transform(data, widths=np.arange(1, 31)):
    transformed_data = {}
    for key, signal in data.items():
        coeffs = cwt(signal, ricker, widths)
        transformed_data[key] = coeffs.mean(axis=0)  # Упрощаем: используем среднее значение вместо flatten
    return transformed_data

# Функция для расчёта спектральной плотности мощности (PSD)
def calculate_psd(data, fs=100):
    psd_data = {}
    for key, signal in data.items():
        freqs, psd = welch(signal, fs=fs)
        psd_data[key] = (freqs, psd)
    return psd_data

# Функция для загрузки данных ЭЭГ из .rec файла
def load_eeg_data_from_rec(rec_file):
    """
    Загружает ЭЭГ данные из файла формата .rec.
    """
    f = pyedflib.EdfReader(rec_file)
    n = f.signals_in_file
    signal_labels = f.getSignalLabels()  # Получаем названия сигналов
    signal_data = {label: f.readSignal(i) for i, label in enumerate(signal_labels)}
    f._close()
    return signal_data

# Функция для загрузки гипнограммы из .hyp файла
def load_hypnogram_from_hyp(hyp_file):
    """
    Загружает гипнограмму из файла формата .hyp, игнорируя метаданные.
    """
    hyp_data = []
    with open(hyp_file, 'r') as file:
        for line in file:
            # Пропускаем строки, которые не содержат числовые значения
            try:
                hyp_data.append(int(line.strip()))
            except ValueError:
                continue
    return np.array(hyp_data)

# Вычисление евклидова расстояния между временными рядами
def euclidean(series_a, series_b):
    """
    Вычисляет евклидово расстояние между двумя одномерными временными рядами.
    """
    return np.linalg.norm(series_a - series_b)

# # Вычисление энтропии и относительного содержания информации
# def calculate_entropy(signal):
#     """
#     Рассчитывает энтропию сигнала на основе гистограммы.
#     """
#     hist, _ = np.histogram(signal, bins=64, density=True)
#     hist = np.clip(hist, 1e-12, None)  # Исключаем нули для логарифма
#     entropy = -np.sum(hist * np.log(hist))
#     return entropy

# Метод для вычисления энтропии для данных сигналов
def calc_entropy(signals):
    """
    Рассчитывает энтропию сигнала на основе гистограммы.
    """
    entropies = []
    for signal in signals:
        # Строим гистограмму для сигнала
        hist, bins = np.histogram(signal, bins='auto', density=True)
        # Рассчитываем энтропию
        H = entropy(hist, base=2)
        entropies.append(H)
    return entropies

# Расчет относительного содержания информации (RE) с учетом минимальных расстояний
def calculate_relative_information(signals, entropies, min_distances):
    """
    Рассчитывает относительное содержание информации (RE) для сигналов.
    """
    # Подсчет энтропий по всем компонентам данного ряда
    all_entropies = calc_entropy(signals)
    total_entropy = np.sum(all_entropies)
    
    # Расчёт (RE) с умножением на дистанцию
    relative_information = [(entropy / total_entropy) * distance 
                            for entropy, distance in zip(all_entropies, min_distances)]
    return relative_information

# Применяем дискретное вейвлет-преобразование Хаара к каждому одномерному ряду
def apply_dwt(signal, flag):
     if flag:
        return signal
     else:
        # Проверяем тип данных и преобразуем в массив, если требуется
        signal = np.asarray(signal)
        # Уровень преобразования адаптивно выбирается на основе длины сигнала
        max_level = pywt.dwt_max_level(signal.size, pywt.Wavelet('haar').dec_len)
        level = min(4, max_level)  # Ограничиваем максимальный уровень до 4
        return pywt.wavedec(signal, 'haar', level=level)[0]


def find_and_remove_closest(wavelet_a, wavelet_b, d_vec):
    if not wavelet_a or not wavelet_b:
        unmatched = wavelet_a | wavelet_b
        return d_vec, unmatched

    # Поиск ближайшей пары сигналов
    (key_a, key_b), min_distance = min(
        ((key_a, key_b), euclidean(signal_a, signal_b))
        for key_a, signal_a in wavelet_a.items()
        for key_b, signal_b in wavelet_b.items()
    )

    # Добавляем минимальное расстояние в список
    d_vec.append(min_distance)

    # Удаляем найденные сигналы
    del wavelet_a[key_a]
    del wavelet_b[key_b]

    # Рекурсивно продолжаем поиск
    return find_and_remove_closest(wavelet_a, wavelet_b, d_vec)

# Нахождение минимальных расстояний для каждой неучтенной компоненты
def find_min_distance_unmatched(unmatched_a, wavelet_b):
    min_distances = []
    for _, signal_a in unmatched_a.items():
        min_distance = min(euclidean(signal_a, signal_b) for signal_b in wavelet_b.values())
        min_distances.append(min_distance)
    return min_distances

# SMETS-анализ
def calculate_smets_distance(model_a, model_b, flag):
    """
    Вычисляет SMETS-расстояние между двумя многомерными временными рядами.

    model_a: dict or np.ndarray
        Первый многомерный временной ряд (словарь с именами сигналов и их значениями) 
    model_b: dict or np.ndarray
        Второй многомерный временной ряд (словарь с именами сигналов и их значениями)
    """
    # Определяем, какая модель имеет большую и меньшую размерность
    len_a = len(model_a.items())
    len_b = len(model_b.items())
    if len_a > len_b:
        m, n  = len_a, len_b
    else:
        m, n  = len_b, len_a

    wavelet_a = {key: apply_dwt(signal,flag) for key, signal in model_a.items()}
    wavelet_b = {key: apply_dwt(signal,flag) for key, signal in model_b.items()}
    
    #  1) Part 1, partial matching
    #  Находим расстояния между самыми ближайшими сигналами (компонентами), которые относятся к разным моделям, 
    #  удаляем их ищем заново, пока в моделе с наименьшей размерностью будут оставаться компоненты
    # 2) Также получаем компоненты unmatched, которым не нашлось пары с минимальным расстоянием
    d_vec, unmatched = find_and_remove_closest(wavelet_a.copy(), wavelet_b.copy(), [])
    
    # Вычисляем p-норму, где p = n (размерность меньшего временного ряда)
    distance = np.linalg.norm(d_vec, ord=n)
    
    if(len_a == len_b):
        return distance
    else:
        # Выбираем какая модель имеет меньшую и большую размерность рядов
        matched, big_with_unmutched = (wavelet_b, wavelet_a)  if len_a > len_b else (wavelet_a, wavelet_b)
        
        # Список, где для каждого неучтенного компонента из unmatched метод находит
        # самое минимальное расстояние до компонент временного ряда другой модели
        min_distances = find_min_distance_unmatched(unmatched, matched)

        # Вычисление этнропии для неучтенных компонентов
        entropies = calc_entropy([signal for _, signal in unmatched.items()])
        
        # Расчет относительного содержания информации (RE) с учетом минимальных расстояний
        # RE= (H_j × min_distance) / ∑H 
        RE = calculate_relative_information([signal for _, signal in big_with_unmutched.items()], entropies, min_distances)

        # Штраф Entropy Penalty EP = сумма(RE_j для всех несопоставимых компонент)
        EP = np.sum(RE)

        # Штраф за несовпадение размерностей
        mismatch_penalty = (m - n) / (m + n) if (m + n) > 0 else 0

        # Итоговая формула SMETS (||d||_n + EP)^2 + P^2
        print(np.sqrt((distance + EP)**2 + mismatch_penalty**2))
        return np.sqrt((distance + EP)**2 + mismatch_penalty**2)


# Функция для вычисления и отображения матриц расстояний между файлами
def compute_file_matrices(all_eeg_data, file_labels):
    """
    Вычисляет матрицы расстояний (евклидова и SMETS) между всеми парами файлов.
    """
    num_files = len(all_eeg_data)
    file_distance_matrix = np.zeros((num_files, num_files))
    file_smets_matrix = np.zeros((num_files, num_files))
    file_smets_matrix_unmod = np.zeros((num_files, num_files))

    for i in tqdm(range(num_files), desc="Calculating matrices"):
        for j in range(i + 1, num_files):
            if i < j:  # Сравниваем только разные файлы (верхний треугольник матрицы)
                # Извлекаем сигналы для двух файлов
                data_i = all_eeg_data[i]
                data_j = all_eeg_data[j]

                # Вычисляем евклидово расстояние между всеми парами сигналов
                signals_i = list(data_i.values())
                signals_j = list(data_j.values())
                
                max_length = max(
                    max(len(signal) for signal in signals_i),
                    max(len(signal) for signal in signals_j),
                )
                
                # Дополняем нулями до максимальной длины
                padded_signals_i = [
                    np.pad(signal, (0, max_length - len(signal)), mode="constant") for signal in signals_i
                ]
                padded_signals_j = [
                    np.pad(signal, (0, max_length - len(signal)), mode="constant") for signal in signals_j
                ]
                
                # Применяем дискретное вейвлет-преобразование к каждому сигналу и сохраняем в словарь
                transformed_data_i = {key: signal for key, signal in zip(data_i.keys(), padded_signals_i)}
                transformed_data_j = {key: signal for key, signal in zip(data_j.keys(), padded_signals_j)}

                # Рассчитываем евклидовы расстояния
                distance_values = []
                for sig_i in padded_signals_i:
                    for sig_j in padded_signals_j:
                        distance_values.append(euclidean(sig_i, sig_j))

                # Среднее евклидово расстояние между файлами
                file_distance_matrix[i, j] = file_distance_matrix[j, i] = np.mean(distance_values)

                # Рассчитываем SMETS-расстояния
                file_smets_matrix[i, j] = file_smets_matrix[j, i] = calculate_smets_distance(transformed_data_i, transformed_data_j, False)

                 # Рассчитываем SMETS-расстояния
                file_smets_matrix_unmod[i, j] = file_smets_matrix_unmod[j, i] = calculate_smets_distance(transformed_data_i, transformed_data_j, True)
                
    file_distance_matrix /= file_distance_matrix.max()
    file_smets_matrix /= file_smets_matrix.max()
    file_smets_matrix_unmod /= file_smets_matrix_unmod.max()
    
    return file_distance_matrix, file_smets_matrix, file_smets_matrix_unmod
```

```python
# Зазгрузка и вывод данных

data_path = "./sleep-edf-database-1.0.0"

# Получаем списки файлов .rec и .hyp
rec_files = [os.path.join(data_path, f) for f in os.listdir(data_path) if f.endswith(".rec")]
hyp_files = [os.path.join(data_path, f) for f in os.listdir(data_path) if f.endswith(".hyp")]

all_eeg_data = []
file_labels = []

for rec_file in tqdm(rec_files, total=len(rec_files), desc="Loading EEG data"):
    eeg_data = load_eeg_data_from_rec(rec_file)

    # Вывод общего графика всех сигналов
    plt.figure(figsize=(12, 6))
    for signal_name, signal_data in eeg_data.items():
        plt.plot(signal_data, label=signal_name)
    plt.title(f"Общий график сигналов из файла {os.path.basename(rec_file)}")
    plt.xlabel("Время")
    plt.ylabel("Амплитуда")
    plt.legend()
    plt.show()

    # Рассчитываем PSD и выводим их на одном графике
    psd_data = calculate_psd(eeg_data)
    plt.figure(figsize=(12, 6))
    for signal_name, (freqs, psd) in psd_data.items():
        plt.semilogy(freqs, psd, label=signal_name)
    plt.title(f"Спектральная плотность мощности (PSD) для файла {os.path.basename(rec_file)}")
    plt.xlabel("Частота (Гц)")
    plt.ylabel("Плотность мощности")
    plt.legend()
    plt.show()

    # Объединение сигналов
    #     concatenated_signals = np.concatenate([v for v in eeg_data.values()])
    #     all_eeg_data.append(concatenated_signals)
    all_eeg_data.append(eeg_data)
    file_labels.append(os.path.basename(rec_file))
```



Если две динамические модели имеют разное количество переменных. В этом случае единственный применяемый метод заключается в определении (взвешенного) среднего поведения каждой модели (группы временных рядов), а затем в сравнении двух средних одномерных временных рядов.

Но может возникнуть необходимость сравнить две группы временных рядов, используя всю содержащуюся в них информацию, но при этом допуская, что в этих двух группах может быть разное количество компонентов.


SMETS подходит для сравнения многомерных временных рядов с разной размерностью. Он вычисляет расстояние между наиболее похожими компонентами двух многомерных временных рядов, а затем добавляет штрафные баллы, чтобы учесть разницу в их размерности. Штрафные баллы рассчитываются с помощью энтропии Шеннона для несовпадающих компонентов. Таким образом, SMETS использует всю информацию, содержащуюся в обоих временных рядах, несмотря на их разную размерность, что делает этот метод уникальным.

Современные методы сравнения многомерных временных рядов, такие как евклидово расстояние, динамическое преобразование времени [16], разложение по сингулярным значениям с взвешенной суммой (WSSVD) [17], коэффициент сходства анализа главных компонент (SPCA) [18] и расширенная норма Фробениуса (EROS) [19], применимы только в тех случаях, когда временные ряды имеют одинаковую размерность. SMETS устраняет это ограничение и позволяет вычислять расстояния даже при разной размерности данных.
